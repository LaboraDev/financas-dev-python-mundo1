```python
# Exerc√≠cio Python 3: Crie um programa que leia dois n√∫meros e mostre a soma entre eles. (Enunciado)
# Criando o programa de leitura com a entrada dos dados sendo capturada pela fun√ß√£o input()
# Esse dado √© do tipo inteiro int()
numero_um = int(input('Digite um n√∫mero inteiro: '))
numero_dois = int(input('Digite outro n√∫mero inteiro: '))
# Aplicada a opera√ß√£o aritm√©tica de adi√ß√£o para o objetivo do exerc√≠cio.
soma = numero_um + numero_dois
# Utilizada a fun√ß√£o print() para exibir na tela o resultado de forma organizada
#com a fun√ß√£o format f'{}'
print(f'A soma entre os n√∫meros {numero_um} e {numero_dois} √© igual a {soma}!')
```

```python
#Exerc√≠cio Python 4: Fa√ßa um programa que leia algo pelo teclado e mostre na tela o seu tipo primitivo e todas as informa√ß√µes poss√≠veis sobre ele. (Enunciado)
# Crinado o programa com a fun√ß√£o input() para realizar a leitura.
dado = input('Digite qualquer coisa: ')
# Exibindo na tela o tipo de informa√ß√£o sobre o dado inserido com a fun√ß√£o print()
print(type(dado)) # Qual o tipo primitivo desse dado? Observa√ß√£o: o retorno sempre ser√° uma string mesmo se digitar n√∫mero(s), pois input sempre retorna str
print(dado.isalnum()) # Esse dado √© um n√∫mero e/ou palavra?
print(dado.isalpha()) # Esse dado a uma palavra?
print(dado.isdecimal()) # Esse dado √© um n√∫mero decimal?
print(dado.islower()) # Esse dado √© composto por letras min√∫sculas?
print(dado.isupper()) # Esse dado √© composto por letras mai√∫sculas?
print(dado.isnumeric()) # Esse dado √© um n√∫mero?
print(dado.isprintable()) # Esse dado √© exibido na tela?
```
